  * `id`

    * ```
    [
    ('Alice', 25), ('Bob', 30), ('Charlie', 35), ('David', 28), ('Eve', 22),

#üí°Ishora: CREATE TABLE, INSERT


````python
#javobni shu yerda yozing.
query = """
CREATE TABLE IF NOT EXISTS users(
id INT,
name TEXT,
birth_date
);
"""
cursor.execute(query)
db.commit()
````

````python
dict = {
    'Alice':25, 
    'Bob':30, 
    'Charlie':35, 
    'David':28, 
    'Eve':22, 
    'Frank':40, 
    'Grace':27, 
    'Hannah':32, 
    'Ivy':29, 
    'Jack':31, 
    'Kevin':33, 
    'Liam':26, 
    'Mia':24, 
    'Noah':36, 
    'Olivia':23, 
    'Peter':39, 
    'Quinn':37, 
    'Rachel':34, 
    'Samuel':41, 
    'Tina':21
}
counter = 1
for key, value in dict.items():
    query = f"INSERT INTO users VALUES ({counter}, '{key}', {value})"
    cursor.execute(query)
    db.commit()
````
